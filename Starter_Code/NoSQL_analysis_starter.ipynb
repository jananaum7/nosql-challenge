### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
results_count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {results_count}")

# Cast the results as a list and save the results to a variable
results = list(establishments.find(query))

# Display the first document in the results using pprint
pprint(results[0])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
#query2 = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': 4}}
query2 = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': 4}}
# Use count_documents to display the number of documents in the result
london_count = establishments.count_documents(query2)
print(f'Number of establishments in London with a rating value of 4 or higher: {london_count}')

# Cast the results as a list and saved the results to a variable 
final_results = list(establishments.find(query2))

# Display the first document in the results using pprint
pprint(final_results)
```

```python
# Convert the results to a Pandas DataFrame
rating_value_df = pd.DataFrame(final_results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(rating_value_df)}")

# Display the first 10 rows of the DataFrame
rating_value_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
penang_flavors = establishments.find_one({"BusinessName": "Penang Flavours"}, 
                                         {"geocode.longitude", "geocode.latitude"}) 
pprint(penang_flavors)
```

```python
# Retrieve a sample document to check the RatingValue field
sample_document = establishments.find_one({}, {"RatingValue": 1})

# Print the document to see the RatingValue and its type
print(f"RatingValue: {sample_document.get('RatingValue')}, Type: {type(sample_document.get('RatingValue'))}")
```

```python
# Set up search parameters
degree_search = 0.01
latitude = float(penang_flavors["geocode"]["latitude"])
longitude = float(penang_flavors["geocode"]["longitude"])

# Create query with string comparisons since coordinates are stored as strings
query = {
    "geocode.latitude": {
        "$gte": str(float(latitude) - degree_search), 
        "$lte": str(float(latitude) + degree_search)
    },
    "geocode.longitude": {
        "$gte": str(float(longitude) - degree_search), 
        "$lte": str(float(longitude) + degree_search)
    },
    "RatingValue": 5
}

# Execute the query and sort by hygiene score
initial_results = establishments.find(query).sort([("scores.hygiene", 1)])
```

```python
# Convert results to Pandas DataFrame
df = pd.DataFrame(list(initial_results))

# Display the results
print(f"Total number of records found: {len(df)}")
print("\nFirst few rows of the DataFrame:")
if not df.empty:
    # Display specific columns for better readability
    columns_to_show = ['BusinessName', 'BusinessType', 'RatingValue', 'geocode', 'AddressLine1']
    print(df[columns_to_show].head())
else:
    print("No results found")

# Display DataFrame information
print("\nDataFrame Information:")
print(df.info())
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Pipeline to:
# 1. Match establishments with a hygiene score of 0
# 2. Group by Local Authority with a count of establishments
# 3. Sort from highest to lowest count of establishments

pipeline_hygiene_0 = [
    # Match stage: Find documents where hygiene score equals 0
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    # Group stage: Group by Local Authority and count the number of establishments
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority Name
            "count": {"$sum": 1}           # Count the number of establishments in each authority
        }
    },
    # Sort stage: Order by count from highest to lowest
    {
        "$sort": {
            "count": -1
        }
    }
]

# Execute pipeline and display results
print("Pipeline: Establishments with hygiene score of 0 grouped by Local Authority")
results_hygiene_0 = list(establishments.aggregate(pipeline_hygiene_0))

# Print the number of documents in the result
print(f"Number of Local Authorities with establishments having a hygiene score of 0: {len(results_hygiene_0)}")

# Print the first 10 results
print("\nFirst 10 results:")
for doc in results_hygiene_0[:10]:
    print(f"Local Authority: {doc['_id']}, Count of Establishments: {doc['count']}")
```
